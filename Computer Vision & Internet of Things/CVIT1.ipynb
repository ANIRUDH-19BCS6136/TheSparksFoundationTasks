# 1. Importing required libraries

```python
import pytesseract
import cv2 as cv
import matplotlib.pyplot as plt
```

# 2. Reading image

```python
image = cv.imread('alphabet.jpg')
plt.imshow(image)
```

# 3. Extracting text from image

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
print(pytesseract.image_to_string('alphabet.jpg'))
```

# 4. Reading handwritten text image

```python
image = cv.imread('handwritten.jpeg')
plt.imshow(image)
```

# 5. Extracting text from handwritten image

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
print(pytesseract.image_to_string('handwritten.jpeg'))
```
